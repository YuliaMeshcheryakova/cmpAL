```python
from sympy import *
```

```python
v = Function('v')
m = 1
g = 1.6
t = symbols("t")
k = 0.01
```

```python
dsolve(diff(v(t), t) + g, v(t))
```

```python
dsolve(diff(v(t), t) + g + k * v(t), v(t))
```

```python
vt = dsolve(diff(v(t), t) + g, v(t), ics = {v(t).subs(t,0):100})
```

```python
vtt = dsolve(diff(v(t), t) + g + k * v(t), v(t), ics = {v(t).subs(t,0):100})
```

```python
vt.rhs
```

```python
vtt.rhs
```

```python
h = Function('h')
ht = dsolve(diff(h(t), t) - vt.rhs, h(t), ics = {h(t).subs(t,0):0})
ht
```

```python
hht = dsolve(diff(h(t), t) - vtt.rhs, h(t), ics = {h(t).subs(t,0):0})
hht
```

```python
solve(vt.rhs, t, 0)
```

```python
time = solve(vtt.rhs, t, 0)
time
```

```python
ht.rhs.subs(t, 62.5)
```

```python
hht.rhs.subs(t, time[0][0])
```
